```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
from IPython.display import display  # <- para ver el DataFrame
import datetime as dt
import numpy as np

df = archivo.copy()

def time_to_seconds(col):
    """
    Convierte una columna que mezcla strings y datetime.time
    a segundos (float).  Devuelve NaN si no se puede parsear.
    """
    def _parse(x):
        if pd.isna(x):
            return np.nan
        # Caso string (e.g. "01:01:41.700000")
        if isinstance(x, str):
            try:
                return pd.to_timedelta(x).total_seconds()
            except ValueError:
                return np.nan
        # Caso datetime.time
        if isinstance(x, dt.time):
            return (x.hour * 3600 +
                    x.minute * 60 +
                    x.second +
                    x.microsecond / 1_000_000)
        # Cualquier otro tipo → NaN
        return np.nan   

    return col.apply(_parse).astype("float")


# GAP numérico
df["GAP"] = pd.to_numeric(df["GAP"], errors="coerce")

# Segundos desde ambas columnas de tiempo
df["Tiempo_sec"]  = time_to_seconds(df["Tiempo"])
df["Tiempo2_sec"] = time_to_seconds(df["Tiempo2"])

# ──────────────────────────────────────────────────────────────
# 3. Ahora tus gráficos funcionarán sin TypeError
# ──────────────────────────────────────────────────────────────

display(df.head())          # Muestra las primeras filas
# print(df.info())          # Si quieres inspeccionar tipos y nulos

# ------------------------------------------------------------------
# 3. Gráficos
# ------------------------------------------------------------------
plt.figure()
plt.boxplot(df["GAP"].dropna())
plt.yscale("log")
plt.title("Distribución de GAP")
plt.ylabel("GAP (escala log)")
plt.tight_layout()
plt.show()

plt.figure()
plt.scatter(df["index"], df["GAP"])
plt.yscale("log")
plt.xlabel("Caso")
plt.ylabel("GAP")
plt.title("GAP por caso (escala log)")
plt.tight_layout()
plt.show()

plt.figure()
df["Mantiene Térmica"].value_counts().plot(kind="bar")
plt.ylabel("Cantidad")
plt.title("Frecuencia de 'Mantiene Térmica'")
plt.tight_layout()
plt.show()

plt.figure()
df["Gap"].value_counts().plot(kind="bar")
plt.ylabel("Cantidad")
plt.title("Frecuencia de 'Gap > 100000'")
plt.tight_layout()
plt.show()

plt.figure()
df["Seteo ganador"].value_counts().plot(kind="bar")
plt.ylabel("Cantidad")
plt.title("Frecuencia de 'Seteo ganador'")
plt.tight_layout()
plt.show()
```

```python
"""
Análisis cuantitativo y gráfico de resultados experimentales (gap2 – tiempos)
Autor: …
Última actualización: 2025-06-30
Requisitos: pandas, numpy, matplotlib, scipy
"""
from __future__ import annotations
import datetime as dt
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, skew, kurtosis

# ------------------------------------------------------------------ #
# 1. Carga de datos y renombrado inicial                             #
# ------------------------------------------------------------------ #
DATA_PATH = Path("resultados_experimentos.xlsx")          # ajusta si lo necesitas
RAW_DF    = archivo.copy()                                # ya lo tienes en memoria
RAW_DF.rename(columns={"GAP": "gap2"}, inplace=True)      # evita colisión con “Gap”

# ------------------------------------------------------------------ #
# 2. Utilidades                                                      #
# ------------------------------------------------------------------ #
def _time_to_seconds(series: pd.Series) -> pd.Series:
    """Convierte strings 'hh:mm:ss', datetime.time o timedelta a segundos (float)."""
    def _parse(x):
        if pd.isna(x):
            return np.nan
        if isinstance(x, dt.timedelta):
            return x.total_seconds()
        if isinstance(x, dt.time):
            return x.hour*3600 + x.minute*60 + x.second + x.microsecond/1_000_000
        if isinstance(x, str):
            try:
                return pd.to_timedelta(x).total_seconds()
            except ValueError:
                return np.nan
        return np.nan
    return series.map(_parse).astype(float)

def prepare_df(df: pd.DataFrame) -> pd.DataFrame:
    # ---- limpieza de cabeceras ----
    df.columns = (
        df.columns.str.strip()
                  .str.lower()
                  .str.replace(" ", "_")
                  .str.normalize("NFKD")          # quita acentos
                  .str.replace(r"[^\w_]", "", regex=True)
    )

    # ---- detectar columna numérica de gap ----
    prefs = ["gap2"] + [c for c in df.columns if c.startswith("gap") and c not in ("gap_100000", "gap")]
    for col in prefs:
        if col in df.columns:
            num = pd.to_numeric(df[col], errors="coerce")
            if num.notna().any():
                df["gap_num"] = num
                break
    else:
        raise ValueError("No se encontró ninguna columna numérica de gap.")

    # ---- tiempos -> segundos ----
    for col in ("tiempo", "tiempo2"):
        if col in df.columns:
            df[f"{col}_sec"] = _time_to_seconds(df[col])

    # ---- índice explícito ----
    if "caso_idx" not in df.columns:
        df.reset_index(drop=False, inplace=True)
        df.rename(columns={"index": "caso_idx"}, inplace=True)

    # ---- variable categórica gap > 100 000 ----
    df["gap_over_100k"] = np.where(df["gap_num"] > 1e5, ">100k", "<=100k")

    return df

df = prepare_df(RAW_DF).copy()

# ------------------------------------------------------------------ #
# 3. Estadística descriptiva                                         #
# ------------------------------------------------------------------ #
num_cols = ["gap_num"] + [c for c in ("tiempo_sec", "tiempo2_sec") if c in df.columns]
descr = df[num_cols].describe(percentiles=[.25, .5, .75]).round(2)

gap_skew, gap_kurt = skew(df["gap_num"].dropna()), kurtosis(df["gap_num"].dropna())
rho, pval = spearmanr(df["gap_num"], df.get("tiempo_sec"), nan_policy="omit")

# ------------------------------------------------------------------ #
# 4. Outliers de gap_num (regla de Tukey)                            #
# ------------------------------------------------------------------ #
q1, q3 = df["gap_num"].quantile([.25, .75])
iqr     = q3 - q1
upper_f = q3 + 1.5 * iqr
outliers = df[df["gap_num"] > upper_f]

# ------------------------------------------------------------------ #
# 5. Visualización                                                   #
# ------------------------------------------------------------------ #
# ---- helper para etiquetar barras ----
def add_bar_labels(ax):
    """Escribe el valor absoluto encima de cada barra."""
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f"{int(height)}",
                    xy=(p.get_x() + p.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha="center", va="bottom", fontsize=8)

fig, axs = plt.subplots(2, 2, figsize=(10, 8), constrained_layout=True)

# 5-A Distribución gap_num
axs[0, 0].boxplot(df["gap_num"].dropna(), vert=True, showfliers=False)
axs[0, 0].set_yscale("log")
axs[0, 0].set_title("Distribución de gap_num (log₁₀)")
axs[0, 0].set_ylabel("gap_num")

# 5-B gap_num por caso
axs[0, 1].scatter(df["caso_idx"], df["gap_num"], alpha=.7)
axs[0, 1].set_yscale("log")
axs[0, 1].set_title("gap_num por caso")
axs[0, 1].set_xlabel("Caso #")
axs[0, 1].set_ylabel("gap_num")

# 5-C Frecuencia Mantiene Térmica
mant_cols = [c for c in df.columns if c.startswith("mantiene")]
if mant_cols:
    counts_mt = df[mant_cols[0]].value_counts().sort_index()
    counts_mt.plot(kind="bar", ax=axs[1, 0])
    add_bar_labels(axs[1, 0])
    axs[1, 0].set_title("Frecuencia: Mantiene Térmica")
    axs[1, 0].set_ylabel("Cuenta")
else:
    axs[1, 0].set_visible(False)

# 5-D gap_num > 100 000
counts_gap = df["gap_over_100k"].value_counts().sort_index()
counts_gap.plot(kind="bar", ax=axs[1, 1])
add_bar_labels(axs[1, 1])
axs[1, 1].set_title("gap_num > 100 000")
axs[1, 1].set_ylabel("Cuenta")

fig.suptitle("Dashboard de resultados experimentales", fontsize=14)
plt.show()

# ------------------------------------------------------------------ #
# 6. Resumen analítico                                               #
# ------------------------------------------------------------------ #
def resumen() -> str:
    linea = "-" * 72
    msg   = [
        "RESUMEN ANALÍTICO".center(72),
        linea,
        f"N° de casos analizados                  : {len(df)}",
        f"gap_num – mediana / media (log)         : "
        f"{df['gap_num'].median():.2f} / {df['gap_num'].mean():.2f}",
        f"Asimetría / Curtosis de gap_num         : {gap_skew:.2f} / {gap_kurt:.2f}",
        f"% casos con gap_num > 100 000           : "
        f"{(df['gap_num'] > 1e5).mean()*100:.1f} %",
        f"N° de outliers (Tukey, 1.5×IQR)         : {len(outliers)}",
        f"Spearman ρ (gap_num vs. tiempo_sec)     : {rho:.3f}   (p={pval:.3g})",
        linea,
    ]
    return "\n".join(msg)

print(resumen())
```
